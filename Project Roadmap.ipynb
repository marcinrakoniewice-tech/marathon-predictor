Marathon Predictor: Pełna Kronika Projektu
1. Przygotowanie Danych i Modelu ML

    Notebook 1: Zarządzanie Danymi – Proces rozpoczął się od umieszczenia surowych zbiorów danych na Digital Ocean Spaces, co zapewniło stabilne źródło danych dla chmury.

    Notebook 2: Pipeline i Modelowanie – Stworzenie kompletnego potoku przetwarzania danych (pipeline), obejmującego odczyt z DO Spaces, trenowanie i walidację.

    Finalizacja Modelu – Wypracowanie ostatecznej wersji modelu uczenia maszynowego i zapisanie go w formacie umożliwiającym produkcyjne wykorzystanie w aplikacji.

2. Ewolucja Aplikacji (app.py)

Proces powstawania interfejsu i logiki AI przebiegał w kilku kluczowych wersjach:

    v1 (GPT-3.5) – Pierwsza działająca wersja wykorzystująca model GPT-3.5 do generowania predykcji.

    v2 (Streamlit) – Wdrożenie interfejsu w bibliotece Streamlit; wersja testowa zidentyfikowała pierwsze błędy w logice wyświetlania.

    v3 (GPT-4o) – Przejście na zaawansowany model GPT-4o, co znacząco poprawiło jakość analiz i stabilność działania Streamlit.

    v4 (Integracja) – Ostateczna wersja rozwojowa z zaimplementowanym modelem GPT-4o i poprawionymi błędami, która stała się fundamentem wersji produkcyjnej.

3. Wdrożenie i DevOps

    Integracja z GitHub – Przeniesienie finalnej wersji kodu (app_v4.py jako główny app.py) do repozytorium GitHub, co umożliwiło automatyzację pracy.

    Konfiguracja Repozytorium – Ustawienie niezbędnych plików konfiguracyjnych, takich jak .python-version (v3.11) oraz requirements.txt ze ściśle określonymi wersjami bibliotek (m.in. pycaret==3.3.2, streamlit==1.32.0).

    Publikacja na Digital Ocean – Finalne wdrożenie aplikacji na Digital Ocean App Platform, konfiguracja zmiennych środowiskowych oraz synchronizacja portów (8080) dla poprawnego działania Health Checks.

4. Monitoring i Analityka (Langfuse)

    Implementacja Langfuse – Podłączenie systemu monitoringu, który loguje każde zapytanie do modelu, pozwala na śledzenie kosztów oraz analizę "ścieżki rozumowania" AI w czasie rzeczywistym.
